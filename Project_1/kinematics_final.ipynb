```python
def stewart_platfornm_func(theta, x1, x2, y2):
    # Define the parameters
    L1 = 2
    L2 = L3 = np.sqrt(2)
    gamma = np.pi/2
    p1 = p2 = p3 = np.sqrt(5)
    
    # Calculate intermediate values
    A2 = L3 * np.cos(theta) - x1
    B2 = L3 * np.sin(theta)
    A3 = L2 * np.cos(theta + gamma) - x2
    B3 = L2 * np.sin(theta + gamma) - y2

    
    # Numerator
    N1 = B3 * (p2 ** 2 - p1 ** 2 - A2 ** 2 - B2 ** 2) - B2 * (p3 ** 2 - p1 ** 2 - A3 ** 2 - B3 ** 2)
    N2 = -A3 * (p2 ** 2 - p1 ** 2 - A2 ** 2 - B2 ** 2) + A2 * (p3 ** 2 - p1 ** 2 - A3 ** 2 - B3 ** 2)

    # Denominator
    D = 2 * (A2 * B3 - B2 * A3)

    # value of x, y to generate function fron p1^2
    # x = N1 / D
    # y = N2 / D
    
    # Equation
    out = N1**2 + N2**2 - p1**2 * D**2
    return out



# Parameter Update x1, x2, y2 value
x1 = 4 
x2 = 0  
y2 = 4     

# Generate theta values
theta_values = np.linspace(-np.pi/4, np.pi/4, 10)
# theta_values = -np.pi/4

# Compute f(theta) values
f_values = stewart_platfornm_func(theta_values, x1, x2, y2)

f_values
```

### Method 1: Newton Raphson Method

```python
# Given parameters
x1 = 5
x2 = 0
y2 = 6
L1 = L3 = 3
L2 = 3 * np.sqrt(2)
gamma = np.pi / 4
p1 = p2 = 5
p3 = 3

# Define the function for Newton's method
def stewart_platform_func(theta):
    A2 = L3 * np.cos(theta) - x1
    B2 = L3 * np.sin(theta)
    A3 = L2 * np.cos(theta + gamma) - x2
    B3 = L2 * np.sin(theta + gamma) - y2

    N1 = B3 * (p2 ** 2 - p1 ** 2 - A2 ** 2 - B2 ** 2) - B2 * (p3 ** 2 - p1 ** 2 - A3 ** 2 - B3 ** 2)
    N2 = -A3 * (p2 ** 2 - p1 ** 2 - A2 ** 2 - B2 ** 2) + A2 * (p3 ** 2 - p1 ** 2 - A3 ** 2 - B3 ** 2)
    D = 2 * (A2 * B3 - B2 * A3)

    # x = N1/D
    # y = N2/D

    return N1**2 + N2**2 - p1**2 * D**2

# function to calculate x and y
def calculate_X_and_Y(theta):
    A2 = L3 * np.cos(theta) - x1
    B2 = L3 * np.sin(theta)
    A3 = L2 * np.cos(theta + gamma) - x2
    B3 = L2 * np.sin(theta + gamma) - y2

    N1 = B3 * (p2 ** 2 - p1 ** 2 - A2 ** 2 - B2 ** 2) - B2 * (p3 ** 2 - p1 ** 2 - A3 ** 2 - B3 ** 2)
    N2 = -A3 * (p2 ** 2 - p1 ** 2 - A2 ** 2 - B2 ** 2) + A2 * (p3 ** 2 - p1 ** 2 - A3 ** 2 - B3 ** 2)
    D = 2 * (A2 * B3 - B2 * A3)
    x = N1/D
    y = N2/D

    return x, y



# Newton's method to find the poses
# Custom Newton's method for finding the root of a function
def custom_newton(func, initial_guess, max_iterations=1000, tolerance=1e-9):
    x = initial_guess
    for _ in range(max_iterations):
        f_x = func(x)
        f_prime_x = (func(x + tolerance) - f_x) / tolerance  # Approximate derivative
        new_x = x - f_x / f_prime_x

        if abs(new_x - x) < tolerance:
            return new_x

        x = new_x

    raise RuntimeError(f"Failed to converge after {max_iterations} iterations")

# Updated function to find the poses using custom Newton's method
def find_poses_custom():
    # Initial guesses for the four poses
    initial_guesses = [-np.pi, np.pi]
    poses = []  # List to store the poses (theta values)

    # Apply custom Newton's method for each initial guess
    for initial_guess in initial_guesses:
        # Use custom Newton's method to find the pose
        pose = custom_newton(stewart_platform_func, initial_guess)
        poses.append(pose)

    return poses

# Find the four poses using custom Newton's method
poses = find_poses_custom()
print("Poses (theta values):", poses)
```

### Method 2: Bisection Method

```python
# Given parameters
x1 = 5
x2 = 0
y2 = 6
L1 = L3 = 3
L2 = 3 * np.sqrt(2)
gamma = np.pi / 4
p1 = p2 = 5
p3 = 3

# Define the function for the Stewart platform
def stewart_platform_func(theta):
    A2 = L3 * np.cos(theta) - x1
    B2 = L3 * np.sin(theta)
    A3 = L2 * np.cos(theta + gamma) - x2
    B3 = L2 * np.sin(theta + gamma) - y2

    N1 = B3 * (p2 ** 2 - p1 ** 2 - A2 ** 2 - B2 ** 2) - B2 * (p3 ** 2 - p1 ** 2 - A3 ** 2 - B3 ** 2)
    N2 = -A3 * (p2 ** 2 - p1 ** 2 - A2 ** 2 - B2 ** 2) + A2 * (p3 ** 2 - p1 ** 2 - A3 ** 2 - B3 ** 2)
    D = 2 * (A2 * B3 - B2 * A3)

    return N1**2 + N2**2 - p1**2 * D**2

# Bisection method to find the poses
def find_poses_bisection():
    num_poses = 2  # We want to find exactly 2 poses

    # List to store the poses (theta values)
    poses = []

    # Iterate to find the specified number of poses
    for _ in range(num_poses):
        # Initial guesses for the bisection method
        left_guess = -np.pi
        right_guess = np.pi

        # Bisection loop
        for _ in range(1000):  # Maximum iterations
            mid_guess = (left_guess + right_guess) / 2
            f_mid = stewart_platform_func(mid_guess)

            if np.abs(f_mid) < 1e-9:
                # Root found
                pose = mid_guess
                poses.append(pose)
                break

            # Update the interval for the next iteration
            if stewart_platform_func(left_guess) * f_mid < 0:
                right_guess = mid_guess
            else:
                left_guess = mid_guess

        else:
            raise RuntimeError("Bisection method did not converge within the maximum number of iterations.")

    return poses

# Find the four poses using the bisection method
poses_bisection = find_poses_bisection()
print("Poses using the bisection method (theta values):", poses_bisection)
```

```python
# Generate theta values and compute f(theta) values for plotting
theta_values = np.linspace(-np.pi, np.pi, 200)
f_values = stewart_platform_func(theta_values)

# Function to find the roots of stewart_platform_func(theta)
def find_roots():
    # Generate theta values
    theta_values = np.linspace(-np.pi, np.pi, 500)
    f_values = stewart_platform_func(theta_values)

    # Find indices where the graph intersects y=0
    zero_crossings = np.where(np.diff(np.sign(f_values)))[0]

    # Interpolate to find precise root values
    roots = []
    for idx in zero_crossings:
        root = np.interp(0, [f_values[idx], f_values[idx + 1]], [theta_values[idx], theta_values[idx + 1]])
        roots.append(root)

    return roots

# Find the roots of the function
roots2 = find_roots()

# Plot the f(theta) curve and the roots
plt.figure(figsize=(10, 6))
plt.plot(theta_values, f_values, label='f(θ)')
plt.axhline(0, color='black', linewidth=0.7, linestyle='--')
plt.plot(roots2, np.zeros_like(roots2), 'ro', label='Roots')

for i, root in enumerate(roots2):
    plt.annotate(f'R_{i+1}', (root, 0), textcoords="offset points", xytext=(0, 10), ha='center')

plt.xlabel('Theta')
plt.ylabel('f(theta)')
plt.title('Plot of f(theta) with Roots')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Function to calculate the length of a strut given its coordinates (x1, y1) and (x2, y2)
def calculate_strut_length(x1, y1, x2, y2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Given lengths
given_p1 = p1
given_p2 = p2
given_p3 = p3

# Calculate lengths of the struts for each pose
lengths_pose = []
for i, theta in enumerate(theta_values):
    # Compute vertices for the pose
    x_A, y_A, x_B, y_B, x_C, y_C = calculate_vertices(theta)

    # Calculate strut lengths for this pose
    length_strut1 = calculate_strut_length(x_A, y_A, 0, 0)
    length_strut2 = calculate_strut_length(x_B, y_B, x1, 0)
    length_strut3 = calculate_strut_length(x_C, y_C, x2, y2)

    # Round off the lengths to the nearest integer
    length_strut1 = round(length_strut1)
    length_strut2 = round(length_strut2)
    length_strut3 = round(length_strut3)

    # Append the lengths for this pose
    lengths_pose.append((length_strut1, length_strut2, length_strut3))

    # Check if the lengths match the given lengths for p1, p2, p3
    is_p1_match = np.isclose(length_strut1, given_p1, rtol=1e-3)
    is_p2_match = np.isclose(length_strut2, given_p2, rtol=1e-3)
    is_p3_match = np.isclose(length_strut3, given_p3, rtol=1e-3)

    print(f"Pose {i+1} (theta ≈ {theta:.4f}):")
    print("Strut 1, p1 match:", is_p1_match)
    print("Strut 2, p2 match:", is_p2_match)
    print("Strut 3, p3 match:", is_p3_match)
    print(f"Strut 1 length: {int(length_strut1)}")
    print(f"Strut 2 length: {int(length_strut2)}")
    print(f"Strut 3 length: {int(length_strut3)}")
    print()  # Empty line between poses
    
# Print all lengths for each pose
print("Strut lengths for each pose:")
for i, (lengths_strut1, lengths_strut2, lengths_strut3) in enumerate(lengths_pose):
    print(f"Pose {i+1}: Strut 1 length = {lengths_strut1:.4f}, Strut 2 length = {lengths_strut2:.4f}, Strut 3 length = {lengths_strut3:.4f}")
```
